# Stats of Call Logs
## 1. accessing logs in xml format

```python
import xml.etree.ElementTree as ET
import pandas as pd


xml_data = open('calllogs.xml', 'r').read()  # Read file
root = ET.XML(xml_data)  # Parse XML

data = []
cols = []
for i, child in enumerate(root):
    data.append([subchild.text for subchild in child])
    cols.append(child.tag)

df = pd.DataFrame(data).T  # Write in DF and transpose it
df.columns = cols  # Update column names

df = df.T
df.columns = ['Phone number', 'Time','Call Duration', 'Log Type']
df.index = list(range(3000))

df['Call Duration'] = pd.to_numeric(df['Call Duration'])
df
```

```python
import tabula

df = tabula.read_pdf('CL.pdf',pages='all')
```

## Concatenating all Pages

```python
for i in range(len(df)):
    #print(df[i])
    df[0] = pd.concat([df[0], df[i]])
df[0]
df2 = df[0] #making a Fresh Dataframe

df2 = df2.drop_duplicates() # Dropping Repeated Values
df2.index = list(range(len(df2))) # Redefining the Index Column
df2['Time']= pd.to_datetime(df2['Time']) # Changing Data Type
df2['Duration'] = pd.to_datetime(df2['Duration'],format= '%H:%M:%S').dt.time
df2['seconds'] = df2['Duration'].apply(lambda x : x.hour * 60 * 60 + x.minute * 60  + x.second)
```

# Problems to Solve
## 1. Extract duration
## 2. make bins for time in a day.
## 3. Find Whats causing difference in number of rows in two dataframes
## 4. Extract Numer of rows where Duration is 0 and type, Dialled

```python
def AllCallStats(df):
    Stats = pd.DataFrame(columns = ['Name', 'Total Calls','TC I Dialed','Received','Missed','Rejected','Total time on Call'])
    #Make Logic for This!
    for name in df2['Name'].value_counts().index:
        newdf = df[df['Name'] == name]
        ################
        try:
            d = newdf['Type'].value_counts()['Dialed']
        except:
            d = 0
        ################
        try:
            rc = newdf['Type'].value_counts()['Received']
        except:
            rc = 0
            
        try:
            m = newdf['Type'].value_counts()['Missed']
        except:
            m = 0   
        
        try:
            rj = newdf['Type'].value_counts()['Rejected']
        except:
            rj = 0
            
            
            
        
            
        Stats.loc[len(Stats.index)] = [name,
                                       len(newdf),
                                       d,
                                       rc,
                                       m,
                                       rj,
                                       sum(newdf['seconds'])/60]
        
    
    return Stats
```
